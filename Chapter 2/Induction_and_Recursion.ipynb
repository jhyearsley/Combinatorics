### 2.1.2 Binomial Coefficients and the Binomial Theorem
#### 73 Use the fact that $(x+y)^n = (x+y)(x+y)^{n-1}$ to give an inductive proof of the binomial theorem.
The claim is that $(x+y)^n = \sum_{i=0}^n \binom{n}{i}x^i y^{n-i}$ for each nonnegative integer $n$. The base case is for $n=0$ and in this case the relation holds since
$$(x+y)^0 = 1 = \sum_{i=0}^0 \binom{0}{i} x^i y^{0-i} 1.$$

Now assume the relation holds for some integer $n-1$ and we prove that it holds for $n$:
$$\begin{align*}
(x+y)^n = (x+y)(x+y)^{n-1} &= (x+y) \sum_{i=0}^{n-1}\binom{n-1}{i} x^i y^{n-1-i} \\
&= \sum_{i=0}^{n-1}\binom{n-1}{i} x^{i+1} y^{n-1-i} + \sum_{i=0}^{n-1}\binom{n-1}{i} x^i y^{n-i} \quad \text{(distribute $x$ and $y$)}\\
&= \sum_{i=1}^{n} \binom{n-1}{i-1}x^i y^{n-i} + \sum_{i=0}^{n-1}\binom{n-1}{i}x^i y^{n-i} \quad \text{(shift index of first sum)}\\
&= \sum_{i=1}^{n-1} \left(\binom{n-1}{i-1} + \binom{n-1}{i}\right)x^i y^{n-1} + \binom{n-1}{n-1}x^ny^0 + \binom{n-1}{0}x^0y^n \quad \text{(rewrite sums and factor)}\\
&= \sum_{i=1}^{n-1}\binom{n}{i}x^i y^{n-i} + \binom{n}{n}x^n y^{n-n} + \binom{n}{0}x^0y^n \quad \text{(pascal identity and rewrite coefs of last two terms)}\\
&= \sum_{i=0}^n \binom{n}{i}x^i y^{n-i} \quad \text{(combine into one sum)}
\end{align*}$$

#### 74 Suppose that $f$ is a function defined on the nonnegative integers such that $f(0) = 3$ and $f(n) = 2f(n − 1)$. Find a formula for $f(n)$ and prove your formula is correct.
Claim: $f(n) = 3\cdot 2^n$ for $n=0,1,2\cdots$. Proof is done by induction on $n$. The base case is $n=1$ which holds since
$$f(1) = 3\cdot 2 = 6 = 2\cdot 3 \cdot 2^0 = 2f(0).$$
Now inductively assume that $f(n-1)=3\cdot 2^{n-1}$ so that the hypothesis holds for some integer $n-1$. Then we have
$$f(n) = 2 f(n-1) = 2\cdot 3\cdot 2^{n-1} = 3\cdot 2^{n}$$

### 2.2 Recurrence Relations
#### 87 How is the number of subsets of an $n$-element set related to the number of subsets of an $(n − 1)$-element set?
The number of $n$ element subsets is equal to the number of $(n-1)$-element subsets multiplied by $2$. The reasoning is that if we take a set $S$ with $(n-1)$-elements, from previous problems we know that there are $2^{n-1}$ subsets. Denote the set of subsets as $P_S$. If we then add one more element to $S$, say $x$, and we want to construct the new set of subsets $P_{S'}$, then we must either add or withold $x$ from each of the sets in $P_S$. In other words, $P_{S'}$ will contain all of the same elements of $P_S$, plus all of the sets from $P_S$ with $x$ added in. 

The recurrence relation is thus $S_n = 2 S_{n-1}$, where $S_0=1$. The claim is that $S_n = 2^{n}$ and the proof is done by induction on $n$. The base case is $n=1$ which holds since a set with $1$ elements has, $2^1$ subsets -- one set containing the element and the other being the empty set. Now inductively assume that $S_{n-1} = 2^{n-1}$ so that the relation holds for some integer $n-1$. Then we have
$$S_n = 2 \cdot S_{n-1} = 2 \cdot 2^{n-1} = 2^n.$$

#### 88  Explain why it is that the number of bijections from an $n$-element set to an $n$-element set is equal to $n$ times the number of bijections from an $(n − 1)$-element subset to an $(n − 1)$-element set.
The number of bijections from an $(n-1)$ element set to an $(n-1)$ element set is equal to $(n-1)!$ which we have shown previously. The reasoning being that the number of bijections is equal to product of the sequence of $n-1$ assignments to be made on the $n-1$ elements, i.e. $n-1$ choices to map the first element to $n-2$ choices to map the second element to and so on and so forth.

Thus the number of bijections from an $n$-element set to an $n$-element set is $n! = n \cdot (n-1)!$. The recurrence relation is given by $B_n = n B_{n-1}$ with $B_0=1$ and the claim is that $B_n = n!$. The proof is done by induction on $n$. The base case holds for $n=1$ since 
$$B_1 = 1\cdot B_0 = 1 = 1!.$$
Now inductively assume that $B_{n-1}=(n-1)!$ for some integer $n-1$. Then we have 
$$B_n = n \cdot B_{n-1} = n \cdot (n-1)! = n!$$

### 2.2.2 Arithmetic Series
#### 93 A child puts away two dollars from her allowance each week. If she starts with twenty dollars, give a recurrence for the amount $a_n$ of money she has after $n$ weeks and find out how much money she has at the end of $n$ weeks.
The recurrence is $a_n = 2 + a_{n-1}$ with $a_0 = 20$ since she start with $20$ dollars. The solution to the recurrence is $a_n = 20 + 2n$. Proof by induction: base case $n=1$ holds since
$$a_1 = 2+a_0 = 2+20 = 22 = 20 + 2\cdot1 = 22.$$
Now inductively assume that $a_{n-1} = 20 + 2(n-1)$ for some integer $n-1$ and we have
$$a_n = 2 + a_{n-1} = 2 + 20 + 2(n-1) = 20 + 2n - 2 + 2 = 20 + 2n$$

#### 94 A sequence that satisfies a recurrence of the form $a_n = a_{n−1} + c$ is called an *arithmetic progression*. Find a formula in terms of the initial value $a_0$ and the common difference $c$ for the term an in an arithmetic progression and prove you are right.
The claim is that $a_n = a_0 + cn$ and the proof is done by induction on $n$. The base case happens when $n=1$ and is true since 
$$a_1 = a_0 + c.$$
Now inductively assume that $a_{n-1} = a_0 + c(n-1)$ for some integer $n-1$ and we have
$$a_n = a_{n-1} + c = a_0 + c(n-1) + c = a_0 + cn - c + c = a_0 + cn.$$

#### 95 A person who is earning \$ $50,000$ per year gets a raise of \$ $3000$ a year for $n$ years in a row. Find a recurrence for the amount $a_n$ of money the person earns over $n + 1$ years. #### 96 An *arithmetic series* is a sequence $s_n$ equal to the sum of the terms $a_0$ through $a_n$ of an arithmetic progression. Find a recurrence for the sum $s_n$ of an arithmetic progression with initial value $a_0$ and common difference $c$ (using the language of Problem 94). Find a formula for general term $s_n$ of an arithmetic series.
The recursion is given by 
$$s_{n} = s_{n-1} + a_n = s_{n-1} + a_0 + nc, \ s_0 = a_0,$$
and the solution is given by 
$$s_n = \frac{(n+1)(a_0 + a_{n})}{2}$$.

The proof is done by induction on $n$ and the base case holds since 
$$
\begin{align*}
s_1 &= 2\frac{(a_0 + a_1)}{2} = a_0 + a_1 \\
s_1 &= s_{0} + a_1 =a_0 + a_1.
\end{align*}$$
Now inductively assume that $s_{n-1} = \frac{n(a_0 + a_{n-1})}{2}$ for some integer $n-1$ and we have that
$$
\begin{align*}
s_n = s_{n-1} + a_n = \frac{n(a_0 + a_{n-1})}{2} + a_n = \frac{n(a_0 + a_0 + (n-1)c)+ 2a_n}{2}  &= \frac{na_0 + na_n -nc+2a_0 + 2nc}{2} \\
&= \frac{na_0 + a_0 + na_n + a_n}{2} \\
&= \frac{(n+1)(a_0 + a_n)}{2}.
\end{align*}$$


### 2.2.3 First Order Linear Recurrences
#### 98 Find a formula in terms of $b$, $d$, $a_0$ and $n$ for the general term $a_n$ of a sequence that satisfies a constant coefficient first order linear recurrence $a_n = ba_{n−1} + d$ and prove you are correct.
The claim is that 
$$a_n = b^n a_0 + \sum_{i=0}^{n-1}d b^i = b^n a_0 + d\frac{1-b^n}{1-b}$$ 
and the proof is done by induction on $n$. The base case holds since 
$$a_1 = b a_0 +d = b^1 a_0 + d\frac{1-b^1}{1-b} = b^1 a_0 + d.$$
Now inductively assume that $a_{n-1} = b^{n-1}a_0 + \sum_{i=0}^{n-2}db^i$ for some integer $n-1$ and then we have that
$$\begin{align*}
a_n = b a_{n-1} + d = b \left(b^{n-1}a_0 + \sum_{i=0}^{n-2}db^i\right) + d =& b^na_0 + \sum_{i=0}^{n-2}db^{i+1} + d \\
&= b^na_0 + \sum_{i=1}^{n-1}db^i + db^0 \quad \text{(shift indices)} \\
&= b^n a_0 + \sum_{i=0}^{n-1}d b^i \quad \text{(add third term into sum)}\\
&= b^n a_0 + d\frac{1-b^{n}}{1-b} \quad \text{(simplify sum of geometric series)}
\end{align*}$$
